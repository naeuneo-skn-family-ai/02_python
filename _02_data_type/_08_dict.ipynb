# dict(Dictionary, 사전 자료형)
- `dict`는 하나의 요소를 Key : Value 형식으로 갖는 자료형이다.
- 사전처럼 단어를 알면 뜻을 알 수 있듯이 Key를 통해 Value를 확인하는 구조
- list처럼 index로 접근하는 것이 아닌 `Key`로 접근 가능
- ---

## 1. dict 생성과 key 규칙

`dict`의 핵심은 `key`이다.

- 하나의 `dict` 안에서 `key`는 중복될 수 없다.
- 같은 `key`를 다시 넣으면 마지막 값으로 덮어써진다.
- `key`는 값을 빠르게 찾기 위한 기준이므로 중간에 값이 바뀌면 안 된다.
- 그래서 `int`, `float`, `str`, `tuple` 같은 불변 자료형은 `key`로 사용할 수 있지만, `list`, `dict`, `set` 같은 가변 자료형은 `key`로 사용할 수 없다.
- Python 3.7 이상에서 `dict`는 삽입 순서를 유지한다.

```python
print('--- dict ---')

dct = {
    'a' : 10,
    'b' : 20,
    'c' : 100
}

print(dct, type(dct))

# key는 불변 타입만 가능
dct2 = {
    10 : 'aaa',
    (10, 20) : 'bbb'
}

print(dct2, type(dct2))


# key 가변 타입 불가 확인
# dct = {
#     [1, 2, 3] : 'ccc'
# }
# TypeError: unhashable type: 'list'
```

```python
print('--- dict 조회 ---')

print("dct :",dct)

# key를 이용한 조회
print("dct['b'] :",dct['b'])
print("dct.get('b') :",dct.get('b'))

# 없는 key 조회
# print("dct['d'] :",dct['d']) # KeyError 발생
print("dct.get('d') :",dct.get('d')) # 없으면 None 반환

#없으면 기본값 적용
print("dct.get('d', '값 없음') :",dct.get('d', '값 없음')) # 값이 없는 경우
print("dct.get('c', '값 없음') :",dct.get('c', '값 없음')) # 값이 있는 경우
```

## dict 요소 추가, 수정, 삭제
- dict는 mutable 자료형
- `dict[없는key] = value` 로 value **추가**
- `dict[중복key] = value` 로 value **수정**
- `dict.update()` 를 이용해 value 수정 가능(여러 값 동시 수정도 가능)
- `pop()` 또는 `del`을 이용해 요소 **제거**

```python
# 추가
dct['e'] = 888
print("추가",dct)

# 수정
dct['e'] = 999
print("수정", dct)

# update()로 수정
dct.update(e = 1234) # Key값 ''생략 -> 알아서 판단한다
print("update()로 수정", dct)

# 여러 value update
dct.update({'a':111, 'b' : 222})
print("여러 value update()", dct)

# 요소 제거
# 1) del
del dct['e']
print("del 연산 :", dct)

# 2) pop() 메서드 == 꺼내기
temp = dct.pop('a')
print(temp)
print(dct)
```

```python
# dict 생성 함수와 주요 메서드
# - dict 생성 방법
# 1) {k : v, k : v, ...}
# 2) dict(key = value, key = value, ..)

dct3 = dict(name = '홍길동', age = 22)
print(dct3)

# 주요 메서드
keys = dct3.keys() # 키만 추출
print(keys, type(keys)) # dict_keys 타입

values = dct3.values() # 값만 추출
print(values, type(values)) # dict_values 타입

items = dct3.items() # item(요소 하나 == k:v 쌍) 추출
print(items, type(items)) # dict_items 타입
```
