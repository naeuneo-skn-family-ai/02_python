# 조건문 if

- 조건문은 특정 조건에 따라 실행할 코드를 선택하는 문법이다.
- `if`는 조건이 참일 때만 실행한다.
- `if...else`는 둘 중 하나를 반드시 실행한다.
- `if...elif...else`는 여러 조건 중 하나를 선택한다.

## 1. if

`if`는 특정 조건에서만 실행되어야 하는 코드가 있을 때 사용한다.

예를 들어 시험 점수가 90점 이상일 때만 커피 쿠폰을 지급할 수 있다.

```python
# give_coupon 함수 선언, 정의
# 매개변수 : int
# 반환 값 : None (없음)

def give_coupon(score: int) -> None:
    if score >= 100:
        print('커피 쿠폰을 드립니다!')


give_coupon(100)  # 함수 호출 (실행)
```

## 2. if...else

`if...else`는 조건이 참인 경우와 거짓인 경우를 모두 처리해야 할 때 사용한다.

예를 들어 점수가 60점 이상이면 합격, 그렇지 않으면 불합격으로 나눌 수 있다.

```python
def pass_or_fail(score: int) -> None :
    if score >= 60:
        print(f"합격입니다. ({score})")
    else:
        print(f"불합격입니다. ({score})")

    print("수고하셨습니다.")
    print("-" * 20)

pass_or_fail(100) # if
pass_or_fail(50) # else
```

## 3. if...elif...else

`if...elif...else`는 고려해야 하는 조건이 여러 개인 경우 사용한다.

학점 계산처럼 여러 구간 중 하나를 선택해야 할 때 적합하다.

```python
def get_grade(score : int) -> str:

    # 유효성 검사 (validation)
    # - 제공 받은 값이 기능 수행에 올바른 값인지 확인

    # 전달 받은 score가 0~100 사이가 아닌 경우
    if not (0 <= score <= 100):
        raise ValueError("0~100 사이 값만 입력해주세요.")

    if score >= 90:
        grade = 'A'
    elif score >= 80:
        grade = 'B'
    elif score >= 70:
        grade = 'C'
    elif score >= 60:
        grade = 'D'
    else:
        grade = 'F'

    return grade

result = get_grade(500)
print(result)
```

## 4. 입력값 유효성 검사

사용자 입력은 항상 올바르다고 가정하면 안 된다. 점수처럼 범위가 정해진 값은 조건문으로 유효성을 먼저 확인하는 습관이 좋다.

```python
score = int(input("점수를 입력하시오 :"))

try: # Error가 발생할 수 있는 구문을 작성해서 시도
    grade = get_grade(score)
    print(f'{score} -> {grade} 등급')
except ValueError as error: # try에서 error 발생 시 수행
    print(error)
```
